```
# Fill category in object site
# database is df, it only for extract category for searching
category_lst = soup.find_all('div', class_='p-filter-item')
# category_lst[0].find_all('a')
category_link_list = []
df = pd.DataFrame()
for i in category_lst:
    category_content = i.span.contents
    print('-'*100)
    print(category_content)
    print('')
    content_arg = []
    for y in i.find_all('a'):
        category_link = y.attrs['href']
        object_content = y.contents[0]
        if 'https://www.hanoicomputer.vn/' in category_link:
            category_link_list.append(category_link)

        elif '//' in category_link:
             category_link_list.append('https://www.hanoicomputer.vn/' + category_link.split('/')[-1])
            #  print(category_link.split('/')[-1])
        elif '/' not in category_link:
            category_link_list.append('https://www.hanoicomputer.vn/' + category_link)
        else: 
            category_link_list.append('https://www.hanoicomputer.vn' + category_link)         
        # print(object_content)
        content_arg.append(object_content)
    # dct_category[category_content[0]] = content_arg
    df_i = pd.DataFrame({category_content[0]: content_arg}) 
    # table of category
    df = pd.concat([df, df_i], axis=1)
print(category_link_list)
print(len(category_link_list))
# category_main = category_lst[0].find_all('span', class_='title js-toggleFilter')
# for lst_category in category_main:
#     print(lst_category.contents)    
```

```
# for each link in category_link_list to find code in each category and compare it with colum code in category_*.csv file
for category_link in category_link_list:
    source = urllib.request.urlopen(category_link)
    category_soup = BeautifulSoup(source)
    # identification arg
    catergory_object_code = []
    obj_code = []
    lst_check_obj = []
    name_category = ''

    # find name of category if '=' in category_link cut it
    if '=' in category_link:
        name_category = category_link.split('=')[-1]
    else:
        name_category = category_link.split('/')[-1]
    
    # count page
    try:
        page_max = category_soup.find('div', class_='paging').find_all('a')[-2].contents[0]
    except:
        page_max = 0
    
    # find code of object in current site
    for page_number in range(int(page_max)+1):
        current_url = category_link + '/'+str(page_number)+'/'
        source = urllib.request.urlopen(current_url)
        currentpage_soup = BeautifulSoup(source)
        lst_objs = currentpage_soup.find_all('div', class_='p-component item')
        # find each code in the lst_objs
        for obj in lst_objs:
            list_p_tag = obj.find_all('p', class_='p-sku')
            for i in list_p_tag:
                if 'Mã SP' in str(i):
                    p_tag = str(i.contents[0]).split(' ')
                    obj_code.append(p_tag[-1]) # return obj_code
    # compare each code in obj_code vs code in category_*.csv file
    # **** edit code in main file, def(Name_main_obj,*)
    df_current_category = pd.read_csv('Category/category_CPU.csv')
    df_code_category = list(df_current_category['Code'])
    # compare each code in df_code_category vs obj_code, if code in df_code_category(code in category_*.csv in the list obj_code: set it's True)
    for i in df_code_category:
        if i in obj_code:
            lst_check_obj.append('True')
        else:
            lst_check_obj.append('False')

    df_current_obj = {name_category: lst_check_obj}
    df_current_obj = pd.DataFrame(df_current_obj)
    df_current_category = pd.concat([df_current_category, df_current_obj], axis=1)
    # df_current_category.to_csv('Category/category_CPU.csv', index=False)

    print('-'*80)
    print(f'category name   : {name_category}')
    print(f'category link   : {category_link}')
    print(f'category code   : {obj_code}')
    print(f'compare code_obj: {lst_check_obj}')
    print(df_current_obj)
    print(df_current_category)
    print(f"{'*'*80}\n")
```
